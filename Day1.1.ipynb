```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.25
b = 0.75
jml_titik = 300

x_point = []
y_point = []

for i in range(jml_titik) :
    x = np.random.normal(0.0,0.4)
    y = a*x + b + np.random.normal(0.0,0.1)
    x_point.append([x])
    y_point.append([y])

plt.scatter(x_point,y_point,label='Random Data')
plt.legend()
plt.show()

np.savetxt('data.csv',np.hstack((x_point,y_point)),fmt='%.5f',delimiter=',')
np.savetxt('data.tsv',np.hstack((x_point,y_point)),fmt='%.5f',delimiter='\t')
```

```python
data = np.genfromtxt('data.csv',delimiter=',')

plt.scatter(data[:,0],data[:,1],label='Data Read')
plt.legend()
plt.show()

# print(data)
```

```python
sumX = np.sum(x_point)
sumY = np.sum(y_point)
sumX2 = np.sum([pow(x_point[i][0],2) for i in range(len(x_point))])
sumXY = np.sum([x_point[i][0]*y_point[i][0] for i in range(len(x_point))])
N = len(x_point)

a = (N*sumXY - sumX*sumY)/((N*sumX2)-pow(sumX,2))
b = (sumY - (a*sumX))/N
print("y = %.2f"%a,"x + %.2f"%b)
```

```python
data = np.genfromtxt('linear-regression.csv',delimiter=',')

plt.scatter(data[:,0],data[:,1],label='Data Read')
plt.legend()
plt.show()

# print(data)
```

```python
sumX = np.sum(data[:,0])
sumY = np.sum(data[:,1])
sumX2 = np.sum([pow(data[i,0],2) for i in range(len(data))])
sumXY = np.sum([data[i,0]*data[i,1] for i in range(len(data))])
N = len(data)

a = (N*sumXY - sumX*sumY)/((N*sumX2)-pow(sumX,2))
b = (sumY - (a*sumX))/N
print("y = %.2f"%a,"x + %.2f"%b)

f = lambda x: a*x + b
x = np.array([min(data[:,0]),max(data[:,0])])

plt.scatter(data[:,0],data[:,1],label='Data Read')
plt.plot(x,f(x),lw=2.5,c='r',label='Regression')
plt.legend()
plt.show()
```
